```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------
# Configure plot fonts and appearance
# --------------------------------------------------
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.size'] = 12

def f(x):
    return np.log(x**4) - 0.7

# --------------------------------------------------
# Domain (exclude zero to avoid log singularity)
# --------------------------------------------------
x = np.linspace(-3, 3, 4000)
x = x[x != 0]
y = f(np.abs(x))

# --------------------------------------------------
# Figure and axes
# --------------------------------------------------
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)

# Explicit axis limits requested
ax.set_xlim(-3, 3)
ax.set_ylim(-25, 10)

# --------------------------------------------------
# Put axes (spines) at x=0 and y=0 so tick labels sit on the axes
# --------------------------------------------------
ax.spines["left"].set_position("zero")
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")

ax.spines["left"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)

ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")

# --------------------------------------------------
# Ticks (FIXED): include positive y ticks up to 10
# --------------------------------------------------
ax.set_xticks(np.arange(-3, 2.9, 0.5))
ax.set_yticks(np.arange(-25, 9, 5))   # <-- fixed (was stopping at 4)

# Grid
ax.minorticks_on()
ax.grid(True, which="both", linestyle="--", linewidth=0.6, alpha=0.7)

# --------------------------------------------------
# Plot the function
# --------------------------------------------------
ax.plot(x, y, linewidth=2, label=r"$f(x)=\ln(x^4)-0.7$")

# --------------------------------------------------
# Positive root marker + annotation
# --------------------------------------------------
root = np.exp(0.175)
ax.scatter(root, 0, s=70, color="green", zorder=5)

ax.annotate(
    rf"Positive root  $x \approx {root:.4f}$",
    xy=(root, 0),
    xytext=(root + 0.35, -6),
    arrowprops=dict(arrowstyle="->", color="red", linewidth=1.5),
    fontsize=12
)

# --------------------------------------------------
# Add arrowheads to positive x and positive y directions
# --------------------------------------------------
xmax = ax.get_xlim()[1]
ymax = ax.get_ylim()[1]

ax.annotate("", xy=(xmax, 0), xytext=(xmax - 0.25, 0),
            arrowprops=dict(arrowstyle="->", color="black", linewidth=1.5))
ax.annotate("", xy=(0, ymax), xytext=(0, ymax - 1.0),
            arrowprops=dict(arrowstyle="->", color="black", linewidth=1.5))

# --------------------------------------------------
# Axis labels near arrow tips (FIXED positions to avoid overlap)
# --------------------------------------------------
# Move "x" slightly right and down so it doesn't overlap with tick label "3"
ax.text(xmax + 0.12, 0, "x", ha="left", va="top", clip_on=False)

# Move "f(x)" slightly right and up for better visibility
ax.text(0, ymax + 0.2, "f(x)", ha="left", va="bottom", clip_on=False)

# --------------------------------------------------
# Legend placed automatically in a free area
# --------------------------------------------------
leg = ax.legend(loc="best", frameon=True)
leg.get_frame().set_alpha(0.85)
leg.get_frame().set_facecolor("white")

plt.tight_layout()
plt.show()
```
